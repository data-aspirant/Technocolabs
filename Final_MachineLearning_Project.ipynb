```python
# read our data
companies_data = pd.read_csv("/content/drive/MyDrive/companies.csv")
companies_data.head()
```

```python
# get the number of missing data points per column
missing_values_count = companies_data.isnull().sum()

# look at the # of missing points in the first ten columns
missing_values_count
```

```python
# how many total missing values do we have?
total_cells = np.product(companies_data.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```

Now we have about 50% of the dataset containing empty cells .
